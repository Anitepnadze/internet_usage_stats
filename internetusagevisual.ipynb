```python
#Visualization code for Internet usage; Case of Georgia


# Importing the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#To change the color of the bars, you can modify the palette parameter. 
#Seaborn provides several predefined palettes, such as 'deep', 'muted', 'bright', 'pastel', 'dark', and 'colorblind'.
#You can also use a list of colors or a Matplotlib colormap.
```

```python
# Data from the article
platform_usage = {'Platform': ['Facebook', 'YouTube', 'Instagram', 'TikTok', 'Odnoklassniki', 'Twitter', 'Vkontakte'],
    'Percentage': [69, 61, 26, 19, 6, 5, 5]}
```

```python
age_group_platforms = {'Age Group': ['18-34', '35-54', '55+'],
    'Average Platforms Used': [3, 2, 1]}
```

```python
settlement_type_platforms = {'Settlement Type': ['Rural', 'Urban', 'Tbilisi'],
    'Average # of Platforms Used': [1, 2, 2]}
```

```python
# Converting data to DataFrames
df_platform_usage = pd.DataFrame(platform_usage)
df_age_group_platforms = pd.DataFrame(age_group_platforms)
df_settlement_type_platforms = pd.DataFrame(settlement_type_platforms)
```

```python
# Plotting the percentage of survey participants using each social media platform
plt.figure(figsize=(10, 6))
sns.barplot(x='Platform', y='Percentage', data=df_platform_usage, palette='viridis')
plt.title('Percentage of Survey Participants Using Each Social Media Platform')
plt.xlabel('Social Media Platform')
plt.ylabel('Percentage')
plt.show()
```

```python
# Plotting the average number of social media platforms used by settlement type
plt.figure(figsize=(10, 6))
sns.barplot(x='Settlement Type', y='Average # of Platforms Used', data=df_settlement_type_platforms, palette='colorblind')
plt.title('Average Number of Social Media Platforms Used by Settlement Type')
plt.xlabel('Settlement Type')
plt.ylabel('Average Number of Platforms Used')
plt.show()
```

```python
#Lineplot for increase in the access to the internet in Georgia 

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {'Date': ['June, 2016', 'June, 2017', 'July, 2018', 'July, 2019', 'July, 2020', 'July, 2021', 'July, 2022', 'July, 2023'],
    'Georgia': [70.1, 70.7, 75.8, 79.3, 83.8, 86.1, 88.4, 89.0],
    'Urban': [79.7, 81.8, 84.2, 86.1, 90.7, 91.4, 92.8, 93.1],
    'Rural': [57.4, 56.2, 64.8, 69.9, 74.5, 78.9, 82.4, 83.4],
    'Samegrelo': [64.1, 62.2, 72.8, 74.0, 77.0, 82.0, 82.1, 84.1]}
```

```python
# Convert the data to a DataFrame
df = pd.DataFrame(data)
```

```python
# Melt the DataFrame to long format
df_melted = pd.melt(df, id_vars=['Date'], var_name='Region', value_name='Percentage')

```

```python
# Plotting the data
plt.figure(figsize=(12, 8))
sns.lineplot(data=df_melted, x='Date', y='Percentage', hue='Region', marker='o')
plt.title('Internet Usage Over Time in Georgia')
plt.xlabel('Date')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.grid(True)
plt.show()
```
